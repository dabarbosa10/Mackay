```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

## Binomial
<br>
The probability distribution to obtain $r$ heads for a bent coin (with prob $f$) that
<br> 
is tossed $N$ times is given by the binomal distribution:
<br>
<br>
$P(r|f,N)=\binom{N}{r}f^r (1-f)^{N-r}$

### Example:
<br>
This example shows the the number of heads distribution  for a bent coin that
<br> 
has $f=0.3$ and is tossed $N=10$.
<br> 
This experiment was repeated 10000.

```python
res=np.random.binomial(10,0.3,10000)
```

```python
plt.hist(res,bins=10,density='True')
plt.show()
```

```python
ax=sns.distplot(res,kde=False)
```

## Problem 2.5
<br>
<br>
In this problem we consider an urn that contains $K$ balls.
<br>
B of these balls are black and the remaining $K-B$ are white ($W$).
<br>

### Question 1:
<br>
The probability of the number of times a black ball is drawn is the Binomial distribution.
<br> 
First, the "probability" of obtaining a black ball is given by:
 <br> 
 <br>
$f_B=\frac{B}{K}$.
<br>
<br>
With this in mind the probability of the number of times is:
<br>
<br>
$P(r|f,N)=\binom{N}{r}f^r (1-f)^{N-r}$

```python
f=2/10
N1=5
N2=400
```

```python
Balls1=np.random.binomial(N1,f,1000);
Balls2=np.random.binomial(N2,f,1000);
```

```python
ax=sns.distplot(Balls1,kde=False)
```

```python
plt.hist(Balls2,density='True')
plt.show()
```

```python
def Mean(N,prob):
    return N*prob
def Var(N,prob):
    return N*f*(1-f)
```

```python
print(Mean(N1,f))
print(Mean(N2,f))
print(Var(N1,f))
print(Var(N2,f))
```
