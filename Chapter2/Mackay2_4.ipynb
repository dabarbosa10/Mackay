```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
```

```python
res=np.random.binomial(10,0.3,10000)
```

```python
plt.hist(res,bins=10,density='True')
plt.show()
```

```python
ax=sns.distplot(res,kde=False)
```

```python
f=2/10
N1=5
N2=400
```

```python
Balls1=np.random.binomial(N1,f,1000);
Balls2=np.random.binomial(N2,f,1000);
```

```python
print(np.mean(Balls1))
print(np.mean(Balls2))
print(np.var(Balls1))
print(np.var(Balls2))
```

```python
ax=sns.distplot(Balls1,kde=False)
```

```python
plt.hist(Balls2,density='True')
plt.show()
```

```python
def Mean(N,prob):
    return N*prob
def Var(N,prob):
    return N*f*(1-f)
```

```python
print(Mean(N1,f))
print(Mean(N2,f))
print(Var(N1,f))
print(Var(N2,f))
```
